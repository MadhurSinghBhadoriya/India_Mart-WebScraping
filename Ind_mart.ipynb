```python
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager

driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())

import pandas as pd
import time
import os
```

```python
driver.get("https://dir.indiamart.com/impcat/disc-harrow.html")
```

```python
storeDetails = driver.find_elements_by_class_name('clg')
# product = driver.find_elements_by_class_name("lg_elps")
names = driver.find_elements_by_class_name('lcname')
contactList = driver.find_elements_by_class_name('pns_h')


nameList = []
# prodList = []
addressList = []
numbersList = []

for i in range(len(names)):
    
    name = names[i].text
    # prod = product[i].text
    address = storeDetails[i].text
    contact = contactList[i].get_attribute("innerHTML")


    nameList.append(name)
    # prodList.append(prod)
    addressList.append(address)
    numbersList.append(contact)


    
# intialise data of lists.
data = {'Company/Brand Name':nameList,
        # 'Product Name':prodList,
        'Address': addressList,
        'Phone':numbersList}

# Create DataFrame
df = pd.DataFrame(data)
```
